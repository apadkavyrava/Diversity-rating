###Calculation of the rating

```
from sklearn.preprocessing import minmax_scale
scal_data = minmax_scale(data)
scaled_data = pd.DataFrame(scal_data)

scaled_data.columns = data.columns

scaled_data.drop('Unnamed: 0', axis=1)
```

```
rate = pd.read_csv('my_sheet.csv - Sheet1 (2).csv')

weights = {}
for index, row in rate.iterrows():
  key = row['first']
  value = row['summ']
  weights[key] = value

def score(row):
  total = 0
  for key in weights.keys():
    total += row[key] * weights[key]
  return total

scores = scaled_data.apply(lambda row: score(row), axis=1)
scaled_data['score'] = scores
sorted_scored = scaled_data.sort_values('score', ascending=False)
sorted_scored['rank'] =sorted_scored['score'].rank(ascending=0)
sorted_scored.head()
```
